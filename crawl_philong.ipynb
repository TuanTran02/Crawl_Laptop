```python
import requests
import re
from requests import get
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

print("Modules imported")
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.implicitly_wait(0)
driver.maximize_window()
print("Driver setup and browser opened")
```

```python
files = open('philong_laptop.txt','r')
lines = files.readlines()
# tạo list trống
list_item = []
for line in lines:
    dict_item = {}
    html = driver.get(line.strip('\n'))
    sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    try:   
        #Lấy tên sản phẩm
        product = soup.find('div', class_='entry-header')
        name = product.find('h1').text
        dict_item['Tên sản phẩm'] = name
        #Lấy giá
        a = soup.find('span', class_='p-price')
        price = a.find('span').text
        dict_item['Giá'] = price
        #Lấy màn hình
        b = soup.find('div', class_='pro-info')
        c = b.find_all('p')
        monitor = c[4].text
        monitor = monitor.replace('Màn hình:', '')
        monitor = monitor.replace('Màn Hình:', '')
        dict_item['Màn hình'] = monitor
        #Lấy CPU
        CPU = c[0].text
        CPU = CPU.replace('CPU:', '')
        CPU = CPU.replace('Cpu:', '')
        CPU = CPU.replace('Bộ vi xử lý:', '')
        dict_item['CPU'] = CPU
        #Lấy RAM
        RAM = c[1].text
        RAM = RAM.replace('RAM:', '')
        RAM = RAM.replace('Ram:', '')
        RAM = RAM.replace('Ram', '')
        dict_item['RAM'] = RAM
        list_item.append(dict_item)
    except:
        dict_item['Tên sản phẩm'] = None
        dict_item['Giá'] = None
        dict_item['Màn hình'] = None
        dict_item['CPU'] = None
        dict_item['RAM'] = None
        list_item.append(dict_item)
```

```python
import pandas as pd
df = pd.DataFrame(list_item)
df
```

```python
df.to_csv('philong_laptop.csv')
```
