```python
import requests
import re
from requests import get
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

print("Modules imported")
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.implicitly_wait(0)
driver.maximize_window()
print("Driver setup and browser opened")
```

```python
files = open('fptshop_phones.txt','r')
lines = files.readlines()
list_item = []
for line in lines:
    dict_item = {}
    html = driver.get(line.strip('\n'))
    sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    # Lấy tên 
    name = soup.find('div', class_='l-pd-top').find('h1').text.split('(',)[0].strip()
    dict_item['Tên sản phẩm'] = name
    # Lấy giá
    price = int(soup.find('div', class_='l-pd-right').find('div', 'st-price-main').text.replace('₫', '').replace('.',''))
    dict_item['Giá'] = price
    try:
        a = soup.find('div', class_='st-param')
        b = a.find_all('li')
        # Lấy màn hình
        monitor = b[0].text
        dict_item['Màn hình'] = monitor
        # Lấy camera sau
        behind_camera = b[1].text
        dict_item['Camera sau'] = behind_camera
        # Lấy camera trước
        front_camera = b[2].text
        dict_item['Camera trước'] = front_camera
        # Chip
        cpu = b[3].text
        dict_item['Chip'] = cpu
        # Bộ nhớ
        ram = b[4].text
        dict_item['Bộ nhớ'] = ram
        list_item.append(dict_item)
    except:
        a = soup.find('div', class_='st-param')
        b = a.find_all('li')
        # Lấy màn hình
        monitor = b[0].text
        dict_item['Màn hình'] = monitor
        # Lấy camera sau
        behind_camera = b[1].text
        dict_item['Camera sau'] = behind_camera
        # Lấy camera trước
        front_camera = b[2].text
        dict_item['Camera trước'] = front_camera
        # Lấy chip
        dict_item['Chip'] = None
        # Bộ nhớ
        ram = b[3].text
        dict_item['Bộ nhớ'] = ram
        list_item.append(dict_item)
```
