```python
files = open('pv_laptop.txt','r')
lines = files.readlines()
# tạo list trống
list_item = []
for line in lines:
    dict_item = {}
    html = driver.get(line.strip('\n'))
    sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    try:
        #Lấy tên sản phẩm
        name = soup.find('div',class_='css-6b3ezu').find('h1').text
        dict_item['Tên'] =name
    except:
        dict_item['Tên'] = None
    #Lấy giá
    try:
        price = soup.find('div',class_='css-2zf5gn').find('div',class_ = 'att-product-detail-latest-price css-roachw').text
        dict_item['Giá'] = price
    except:
        dict_item['Giá'] = None
    #Lấy info
    try:
        b = soup.find('div', class_='css-1i1dodm')
        c = b.find('div', class_='css-1nv5d5l')
        if c.find('b'):
            lines = c.contents
            #Lấy màn hình
            man_hinh = c.contents[4].strip().split(': ')[1]
            dict_item['Màn hình']=man_hinh
            #Lấy CPU
            cpu = c.contents[2].strip().split(': ')[1]
            dict_item['CPU']=cpu
            #Lấy RAM
            ram = c.contents[6].strip().split(': ')[1]
            dict_item['RAM']=ram
            list_item.append(dict_item)
        else:
            lines = c.contents
            #Lấy màn hình
            man_hinh = c.contents[2].strip().split(': ')[1]
            dict_item['Màn hình']=man_hinh
            #Lấy CPU
            cpu = c.contents[0].strip().split(': ')[1]
            dict_item['CPU']=cpu
            #Lấy RAM
            ram = c.contents[4].strip().split(': ')[1]
            dict_item['RAM']=ram
            list_item.append(dict_item)
    except:
        b = soup.find('div', class_='css-1i1dodm')
        c = b.find('div', class_='css-1nv5d5l')
        d = c.find('p')
        lines = d.contents
        man_hinh = d.contents[2].strip().split(': ')[1]
        dict_item['Màn hình']=man_hinh
        #Lấy CPU
        cpu = d.contents[0].strip().split(': ')[1]
        dict_item['CPU']=cpu
        #Lấy RAM
        ram = d.contents[4].strip().split(': ')[1]
        dict_item['RAM']=ram
        list_item.append(dict_item)
```

```python
import pandas as pd
df = pd.DataFrame(list_item)
df
```

```python
df.to_csv('pv_laptop1.csv')
```

```python
import pandas as pd
import glob

# Đường dẫn đến thư mục chứa các tệp CSV cần gộp
folder_path = 'C:\\Python\\Crawl_Web\\Data_squad'

# Đọc tất cả các tệp CSV trong thư mục
csv_files = glob.glob(folder_path + '/*.csv')

# Tạo một danh sách để lưu trữ các DataFrame từ các tệp CSV
dataframes = []

# Đọc từng tệp CSV và thêm vào danh sách DataFrame
for file in csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Gộp các DataFrame thành một DataFrame duy nhất
merged_df = pd.concat(dataframes)

# Lưu DataFrame kết quả thành tệp CSV
merged_df.to_csv('C:\Python\Crawl_Web\Data_squad\Laptop_Final.csv', index=False)
```
